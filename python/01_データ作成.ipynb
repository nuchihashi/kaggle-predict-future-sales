```python
import numpy as np
import pandas as pd
import datetime
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 100)
```

```python
items = pd.read_csv('..\\input_data\\items_en.csv')
shops = pd.read_csv('..\\input_data\\shops_en.csv')
item_categories = pd.read_csv('..\\input_data\\item_categories_en.csv')
sales_train = pd.read_csv('..\\input_data\\sales_train.csv')
# set index to ID to avoid droping it later
test  = pd.read_csv('..\\input_data\\test.csv').set_index('ID')
```

```python
items_with_categories = pd.merge(items, item_categories, on="item_category_id")
sales_with_items = pd.merge(sales_train, items_with_categories, on="item_id")
train_merged = pd.merge(sales_with_items, shops, on="shop_id")
train_merged.info()
```

```python
train_merged['date2'] = train_merged['date'].apply(lambda x:datetime.datetime.strptime(x, '%d.%m.%Y'))
```

```python
train_merged.isnull().sum()
```

```python
train_merged
```

```python
result=train_merged.groupby(['date_block_num','date2']).count().to_html()
```

```python
text_file = open("..\\result\\01_check_values\\01_date.html", "w")
text_file.write(result)
text_file.close()
```

```python
train_merged['year'] = train_merged['date2'].apply(lambda x: x.year)
train_merged['month'] = train_merged['date2'].apply(lambda x: x.month)
train_merged['day'] = train_merged['date2'].apply(lambda x: x.day)
train_merged['weekday'] = train_merged['date2'].apply(lambda x: x.isoweekday())
train_merged['revenue'] = train_merged['item_price'] * train_merged['item_cnt_day']
```
